```python

sns.heatmap(train.iloc[:,1:].corr(), vmax=1 ,vmin=-1 ,square=True,annot=True)
```

Heatmap - this is correlation matrix . positive correlation between fare and survived and fare and age
 Parch (number of parents/children aboard) and survived 
Pclass and Fare have a notable negative correlation of -0.55, which could mean that as the passenger class (where 1st class is higher) increases, the fare tends to decrease.
1 indicates a perfect positive correlation: as one variable increases, the other variable also increases proportionally.
-1 indicates a perfect negative correlation: as one variable increases, the other variable decreases proportionally.
0 indicates no correlation: the variables are independent of each other.
A negative correlation of -0.34 between Pclass and Survived means that as the Pclass value increases, the Survived value tends to decrease, and vice versa.
